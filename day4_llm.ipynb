```python
from pypdf import PdfReader
```

```python
def load_pdf(path):
    reader = PdfReader(path)
    text=""
    for page in reader.pages:
        text+=page.extract_text() + "\n"
    return text
```

```python
text = load_pdf("C:/Users/amrit/OneDrive/Desktop/Self Learning/Attention is all you need.pdf")
```

```python
def chunk_text(text, chunk_size=300, overlap=50):
    chunks=[]
    start=0

    while start<len(text):
        end = start+chunk_size
        chunks.append(text[start:end])
        start+=chunk_size+overlap
    return chunks
```

```python
chunks = chunk_text(text)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

embedder = SentenceTransformer("all-MiniLM-L6-v2")

def create_index(chunks):
    embeddings = embedder.encode(chunks)
    return embeddings

def retrieve(query, chunks, embeddings, top_k=3):
    query_embedding = embedder.encode(query)
    scores = cosine_similarity([query_embedding],embeddings)[0]
    top_indices = np.argsort(scores)[-top_k:][::-1]
    return [chunks[i] for i in top_indices]
```

```python
text = load_pdf("Manufacturing_Safety_Instructions.pdf")
chunks = chunk_text(text)
embeddings = create_index(chunks)
retrieved_chunks = retrieve("What should be reported to the supervisor?",chunks,embeddings)
```
